```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import matplotlib.patches as mpatches
import matplotlib.collections as mcollections
```

```python
df = pd.read_csv('farpod20.csv')
print(df)
df1 = df
data = [df['lgt_pct'],df['mod_pct'],df['hvy_pct']]
plt.violinplot(data, showmedians=True)
plt.xticks([1, 2, 3], ['Light', 'Moderate', 'Heavy'])
plt.ylim(0, 100)
plt.xlabel('Snowfall Intensity')
plt.ylabel('Percentage of Snowfall Intensity')
plt.title('ASOS/AWOS Snowfall Intensity Distribution for 2019-2020')
plt.grid(axis='y')
```

```python
labels = ['No Snow', 'Snow']
far = (df['far'].sum() / (df['far'].sum() + df['pod'].sum()))
pod = (df['pod'].sum() / (df['far'].sum() + df['pod'].sum()))
percentages = [df['far_pct'].mean(), df['pod_pct'].mean()]
percentages = [far, pod]
plt.pie(percentages, labels=labels, autopct='%1.1f%%', startangle=90, colors=['indianred', 'lightgreen'],
        wedgeprops={'edgecolor':'white'})
plt.title('ASOS/AWOS Snow and No Snow Observations for 2019-2020')
```

```python
data = [df1['lgt_pct'], df2['lgt_pct'], df4['lgt_pct']]
years = ['2019-2020', '2020-2021', '2022-2023']
plt.violinplot(data, showmedians=True)
plt.xticks([1, 2, 3], ['2019-2020', '2020-2021', '2022-2023'])
plt.ylim(0, 100)
plt.xlabel('Years')
plt.ylabel('Percentage of Snowfall Intensity')
plt.title('ASOS/AWOS Light Snowfall Intensity Distribution for 2019-2023')
plt.grid(axis='y')

X = np.arange(1, len(years) + 1)
y = np.array([np.mean(d) for d in data])

slope, intercept = np.polyfit(X, y, deg=1)
# Plot trend line (linear fit)
trend_line, = plt.plot(X, slope * X + intercept, marker='o', linestyle='-', color='r', label='Linear Fit Trend')
# Annotate the equation of the trend line on the plot
equation_text = f'Linear Fit: y = {slope:.2f}x + {intercept:.2f}'
plt.annotate(equation_text, xy=(0.5, 0.05), xycoords='axes fraction', ha='center', fontsize=10, color='r',weight='bold')
```

```python
df3 = pd.read_csv('noahfarpod.csv')
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(24, 8))

# List of DataFrames
dfs = [df1, df2, df3, df4]

# List of years
years = ['2019-2020', '2020-2021', '2021-2022', '2022-2023']

# List of colors for each year
colors = ['blue', 'orange', 'green', 'red']

# Iterate over each intensity category
for ax, intensity in zip(axes, ['lgt', 'mod', 'hvy']):
    data = [df[intensity + '_pct'] for df in dfs]

    # Create positions for each year
    positions = [i + 1 for i in range(len(years))]
    
    # Iterate over each year
    for i, (year, color) in enumerate(zip(years, colors)):
        parts = ax.violinplot(data[i], positions=[positions[i]], showmedians=True, showextrema=True, widths=0.7)

        # Set color for the violin plot
        for pc in parts['bodies']:
            pc.set_facecolor(color)

# Customize plots
for i, ax in enumerate(axes):
    ax.set_xticks([1, 2, 3, 4])
    ax.set_xticklabels(years)
    ax.set_ylim(0, 100)
    ax.set_xlabel('Years', fontsize=14)
    ax.set_ylabel('Percentage of Observations Per Event', fontsize=14)
    ax.set_title(f' {["Light", "Moderate", "Heavy"][i]} Snowfall Intensity', fontsize=16)
    ax.grid(axis='y')

# Add an overall title
fig.suptitle('Distribution of ASOS/AWOS Snowfall Intensity Observations for 2019-2023', fontsize=18)

# Add legend
legend_handles = [Patch(color=color, label=year) for year, color in zip(years, colors)]
#fig.legend(handles=legend_handles, title='Years', loc='upper right')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18, 6))

# List of DataFrames
dfs = [df1, df2, df4]  # Assuming df3 is not needed

# List of years
years = ['2019-2020', '2020-2021', '2021-2022']

# List of colors for each year
colors = ['blue', 'orange', 'green']

# List to store violin plot parts for each year
violin_parts = []

for i, (df, year, color) in enumerate(zip(dfs, years, colors)):
    # Plot for each intensity level
    data = [df['lgt_pct'], df['mod_pct'], df['hvy_pct']]
    positions = [1 + i, 4 + i, 7 + i]
    parts = ax.violinplot(data, positions=positions, showmedians=True, showextrema=True, widths=0.7)
    
    # Set color for the violin plot
    for pc in parts['bodies']:
        pc.set_facecolor(color)
    
    # Append violin plot parts to the list
    violin_parts.append(parts)
    violin_parts.append(Patch(color=color, label=year))

# Customize plot
ax.set_xticks([2, 5, 8])
ax.set_xticklabels(['Light', 'Moderate', 'Heavy'])
ax.set_ylim(0, 100)
ax.set_xlabel('Snowfall Intensity')
ax.set_ylabel('Percentage of Snowfall Intensity')
ax.set_title('Violin Distribution of ASOS/AWOS Snowfall Intensity Observations for 2019-2022')
ax.grid(axis='y')

# Add legend
ax.legend(violin_parts, years, title='Years')


# Show the plot
plt.show()
```
