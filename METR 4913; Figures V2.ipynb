```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import matplotlib.patches as mpatches
import matplotlib.collections as mcollections
```

```python
df1 = pd.read_csv('farpod20.csv')
df2 = pd.read_csv('cdfarpod.csv')
df3 = pd.read_csv('noahfarpod.csv')
df4 = pd.read_csv('farpod23.csv')
```

```python
data = [df1['lgt_pct'], df2['lgt_pct'], df3['lgt_pct'], df4['lgt_pct']]
years = ['2019-2020', '2020-2021', '2021-2022','2022-2023']
plt.violinplot(data, showmedians=True)
plt.xticks([1, 2, 3, 4], ['2019-2020', '2020-2021', '2021-2022', '2022-2023'])
plt.ylim(0, 100)
plt.xlabel('Years')
plt.ylabel('Percentage of Snowfall Intensity')
plt.title('ASOS/AWOS Light Snowfall Intensity Distribution for 2019-2023')
plt.grid(axis='y')

X = np.arange(1, len(years) + 1)
y = np.array([np.mean(d) for d in data])

slope, intercept = np.polyfit(X, y, deg=1)
# Plot trend line (linear fit)
trend_line, = plt.plot(X, slope * X + intercept, marker='o', linestyle='-', color='r', label='Linear Fit Trend')
# Annotate the equation of the trend line on the plot
equation_text = f'Linear Regression: y = {slope:.2f}x + {intercept:.2f}'
plt.annotate(equation_text, xy=(0.5, 0.05), xycoords='axes fraction', ha='center', fontsize=10, color='r',weight='bold')
```

```python
# Create a 1x3 subplot
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Loop over the data for each intensity level
intensity_levels = ['lgt_pct', 'mod_pct', 'hvy_pct']

for i, intensity in enumerate(intensity_levels):
    data = [df1[intensity], df2[intensity], df3[intensity], df4[intensity]]
    years = ['2019-2020', '2020-2021', '2021-2022', '2022-2023']

    # Violin plot
    axes[i].violinplot(data, showmedians=True)
    axes[i].set_xticks([1, 2, 3, 4])
    axes[i].set_xticklabels(['2019-2020', '2020-2021', '2021-2022', '2022-2023'])
    axes[i].set_ylim(0, 100)
    axes[i].set_xlabel('Years')
    axes[i].set_ylabel('Percentage of Snowfall Intensity')
    axes[i].set_title(f'ASOS/AWOS {intensity.capitalize()} Snowfall Intensity Distribution for 2019-2023')
    axes[i].grid(axis='y')

    # Linear fit
    X = np.arange(1, len(years) + 1)
    y = np.array([np.mean(d) for d in data])
    slope, intercept = np.polyfit(X, y, deg=1)
    trend_line, = axes[i].plot(X, slope * X + intercept, marker='o', linestyle='-', color='r', label='Linear Fit Trend')
    equation_text = f'Linear Regression: y = {slope:.2f}x + {intercept:.2f}'
    axes[i].annotate(equation_text, xy=(0.5, 0.95), xycoords='axes fraction', ha='center', fontsize=10, color='r', weight='bold')

# Adjust layout to prevent clipping of titles
plt.tight_layout()
plt.show()
```
