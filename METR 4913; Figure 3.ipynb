```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import matplotlib.patches as mpatches
import matplotlib.collections as mcollections
```

```python
df1 = pd.read_csv('farpod20.csv')
df2 = pd.read_csv('cdfarpod.csv')
df3 = pd.read_csv('noahfarpod.csv')
df4 = pd.read_csv('farpod23.csv')
```

```python
pod1 = (df1['pod'].sum() / (df1['far'].sum() + df1['pod'].sum()))
pod2 = (df2['pod'].sum() / (df2['far'].sum() + df2['pod'].sum()))
pod3 = (df3['pod'].sum() / (df3['far'].sum() + df3['pod'].sum()))
pod4 = (df4['pod'].sum() / (df4['far'].sum() + df4['pod'].sum()))

intensity_levels = ['lgt_pct', 'mod_pct', 'hvy_pct']
years = ['2019-2020', '2020-2021', '2021-2022', '2022-2023']
X = np.arange(1, len(years) + 1)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

pods = [pod1 * 100, pod2* 100, pod3* 100, pod4* 100]
numeric_years = np.arange(len(years))
slope, intercept = np.polyfit(numeric_years, pods, 1)
equation_text = f'Linear Trend: y = {slope:.2f}x + {intercept:.2f}'
axes[0, 0].scatter(years, pods, color='blue', label='Data Points', zorder=2)
axes[0, 0].plot(years, pods, linestyle='-', color='dodgerblue', zorder=1)
trend_line, = axes[0, 0].plot(numeric_years, slope * numeric_years + intercept, linestyle='--', color='red', label='Linear Trend', linewidth=2)
axes[0, 0].annotate(equation_text, xy=(0.5, 0.955), xycoords='axes fraction', ha='center', fontsize=10, color='red', weight='bold')
axes[0, 0].grid(axis='y')
axes[0, 0].set_yticks(np.arange(65, 76, 2))
axes[0, 0].set_title('Probability of Detection', fontsize=14)
axes[0, 0].set_xlabel('Years', fontsize=12)
axes[0, 0].set_ylabel('Percentage of Observations Per Event', fontsize=12)
axes[0, 0].legend()

data1 = [df1['lgt_pct'], df2['lgt_pct'], df3['lgt_pct'], df4['lgt_pct']]
y1 = np.array([np.mean(d) for d in data1])
axes[0, 1].scatter(X, y1, label='Data Points', color='blue', zorder=2)
axes[0, 1].plot(X, y1, linestyle='-', color='dodgerblue', zorder=1)
axes[0, 1].set_xticks([1, 2, 3, 4])
axes[0, 1].set_xticklabels(['2019-2020', '2020-2021', '2021-2022', '2022-2023'])
axes[0, 1].set_ylim(80, 100)
axes[0, 1].set_yticks(np.arange(80, 101, 5))
axes[0, 1].set_xlabel('Years', fontsize=12)
axes[0, 1].set_ylabel('Percentage of Observations Per Event', fontsize=12)
axes[0, 1].set_title('Light Snowfall Intensity', fontsize=14)
axes[0, 1].grid(axis='y')
slope, intercept = np.polyfit(X, y1, deg=1)
trend_line, = axes[0, 1].plot(X, slope * X + intercept, linestyle='--', color='red', label='Linear Trend', linewidth=2)
equation_text = f'Linear Trend: y = {slope:.2f}x + {intercept:.2f}'
axes[0, 1].annotate(equation_text, xy=(0.5, 0.955), xycoords='axes fraction', ha='center', fontsize=10, color='red', weight='bold')
axes[0, 1].legend()

data2 = [df1['mod_pct'], df2['mod_pct'], df3['mod_pct'], df4['mod_pct']]
y2 = np.array([np.mean(d) for d in data2])
axes[1, 0].scatter(X, y2, label='Data Points', color='blue', zorder=2)
axes[1, 0].plot(X, y2, linestyle='-', color='dodgerblue', zorder=1)
axes[1, 0].set_xticks([1, 2, 3, 4])
axes[1, 0].set_xticklabels(['2019-2020', '2020-2021', '2021-2022', '2022-2023'])
axes[1, 0].set_ylim(0, 20)
axes[1, 0].set_yticks(np.arange(0, 21, 5))
axes[1, 0].set_xlabel('Years', fontsize=12)
axes[1, 0].set_ylabel('Percentage of Observations Per Event', fontsize=12)
axes[1, 0].set_title('Moderate Snowfall Intensity', fontsize=14)
axes[1, 0].grid(axis='y')
slope, intercept = np.polyfit(X, y2, deg=1)
trend_line, = axes[1, 0].plot(X, slope * X + intercept, linestyle='--', color='red', label='Linear Trend', linewidth=2)
equation_text = f'Linear Trend: y = {slope:.2f}x + {intercept:.2f}'
axes[1, 0].annotate(equation_text, xy=(0.5, 0.955), xycoords='axes fraction', ha='center', fontsize=10, color='red', weight='bold')
axes[1, 0].legend(loc='upper left')

data3 = [df1['hvy_pct'], df2['hvy_pct'], df3['hvy_pct'], df4['hvy_pct']]
y3 = np.array([np.mean(d) for d in data3])
axes[1, 1].scatter(X, y3, label='Data Points', color='blue', zorder=2)
axes[1, 1].plot(X, y3, linestyle='-', color='dodgerblue', zorder=1)
axes[1, 1].set_xticks([1, 2, 3, 4])
axes[1, 1].set_xticklabels(['2019-2020', '2020-2021', '2021-2022', '2022-2023'])
axes[1, 1].set_ylim(0, 8)
axes[1, 1].set_yticks(np.arange(0, 9, 2))
axes[1, 1].set_xlabel('Years', fontsize=12)
axes[1, 1].set_ylabel('Percentage of Observations Per Event', fontsize=12)
axes[1, 1].set_title('Heavy Snowfall Intensity', fontsize=14)
axes[1, 1].grid(axis='y')
slope, intercept = np.polyfit(X, y3, deg=1)
trend_line, = axes[1, 1].plot(X, slope * X + intercept, linestyle='--', color='red', label='Linear Trend', linewidth=2)
equation_text = f'Linear Trend: y = {slope:.2f}x + {intercept:.2f}'
axes[1, 1].annotate(equation_text, xy=(0.5, 0.955), xycoords='axes fraction', ha='center', fontsize=10, color='red', weight='bold')
axes[1, 1].legend()

plt.suptitle('Trendlines for Probability of Detection and Snowfall Intensities for 2019-2023', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
